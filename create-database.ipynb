```python
import getpass
import os
import shutil
from dotenv import load_dotenv
from langchain.schema import Document
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import DirectoryLoader, TextLoader
from langchain_text_splitters import MarkdownHeaderTextSplitter
import chromadb
import openai

load_dotenv()

CHROMA_PATH = "chroma"
DATA_PATH = "data"

openai.api_key = os.environ["OPENAI_API_KEY"]

loader = DirectoryLoader(DATA_PATH, glob="**/*.md", loader_cls=TextLoader)
docs = loader.load()



headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
]

splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)

split_documents = []
for doc in docs:
    split_docs = splitter.split_text(doc.page_content)
    for idx, text in enumerate(split_docs):
        split_documents.append({
            "source": doc.metadata.get("source", "unknown"),
            "part": idx + 1,
            "content": text
        })

# for split_doc in split_documents:
#     print(f"Source: {split_doc['source']} - Part {split_doc['part']}")
#     print(split_doc['content'])
#     print("=" * 50)



print(split_documents[0])

if os.path.exists(CHROMA_PATH):
    shutil.rmtree(CHROMA_PATH)

db = Chroma.from_documents(
    documents=split_documents,
    embedding=OpenAIEmbeddings(),
    persist_directory=CHROMA_PATH
)
```
